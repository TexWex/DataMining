```python
list = []

for i in range(len(df)):
    matched = False
    for genre in most_popular:
        if genre in df.iloc[i]['spotify_genre']:
            df.at[i, 'spotify_genre'] = [genre]
            matched = True
            break
    if(not matched):
        list.append(i)
    
df = df.drop(list)

# df = df[len(df['spotify_genre']) == 1]
```

```python
df = df.drop(['song_id', 'performer', 'song', 'spotify_track_id', 'spotify_track_preview_url', 'spotify_track_album', 'mode'], axis=1)
```

## Removing outliers

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# y = df['spotify_genre']
# X = df.drop('spotify_genre', axis=1).dropna()
# display(y)
```

```python
df = df[df['spotify_track_duration_ms'] < 3_000_000]
df = df[df['danceability'] > 0.001]
df = df[df['energy'] > 0.03]
df = df[df['loudness'] < 0]
df = df[df['speechiness'] < 0.8]
df = df[df['tempo'] > 50]
```

```python
mms = MinMaxScaler()

# df['spotify_track_duration_ms'] = mms.fit_transform(df['spotify_track_duration_ms'])
# df['loudness'] = mms.fit_transform(df['loudness'])
# df['tempo'] = mms.fit_transform(df['tempo'])
# df['spotify_track_popularity'] = mms.fit_transform(df['spotify_track_popularity'])

df[['spotify_track_duration_ms','loudness','tempo','spotify_track_popularity']] = mms.fit_transform(df[['spotify_track_duration_ms','loudness', 'tempo', 'spotify_track_popularity']])
```

## Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df['spotify_genre'])
X = df.drop(['spotify_genre', 'spotify_track_explicit', 'time_signature', 'key'], axis=1).dropna()

knn = KNeighborsClassifier(n_neighbors=1, metric="euclidean")

selector = SequentialFeatureSelector(knn, n_features_to_select=5 ,direction="forward")

selector.fit(X, y)

selected_features = selector.get_support()

for i in range(len(selected_features)):
    if selected_features[i]:
        print(X.columns[i])

X_t = selector.transform(X)
```

```python
df = df[df['spotify_track_duration_ms'] < 3_000_000]
df = df[df['danceability'] > 0.001]
df = df[df['energy'] > 0.03]
df = df[df['loudness'] < 0]
df = df[df['speechiness'] < 0.8]
df = df[df['tempo'] > 50]
```

## Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
display(df['spotify_genre'].to_numpy())
y = mlb.fit_transform(df['spotify_genre'].to_numpy())

X = df.drop(['spotify_genre', 'spotify_track_explicit', 'time_signature', 'key'], axis=1).dropna()

knn = KNeighborsClassifier(n_neighbors=1, metric="euclidean")

selector = SequentialFeatureSelector(knn, n_features_to_select=5 ,direction="forward")

selector.fit(X, y)

selected_features = selector.get_support()

for i in range(len(selected_features)):
    if selected_features[i]:
        print(X.columns[i])

X_t = selector.transform(X)
```

# Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelBinarizer
from sklearn.tree import DecisionTreeClassifier

X = X_t
y = df['spotify_genre']
y = y.apply(lambda elem : most_popular.index(elem[0]))
display(y)


# y_dense = LabelBinarizer().fit_transform(y)

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
clf = DecisionTreeClassifier(random_state=0)
scores = cross_val_score(clf, X, y, cv=10)
```
